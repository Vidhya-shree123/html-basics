4: You're creating a form that asks the user to enter their name, select their country from a list, and click a button to submit. Write the HTML code for this.

````python
%%html
<!DOCTYPE html>
<html>
<head>
    <title>User Form</title>
</head>
<body>

    <h2>User Information Form</h2>

    <form>
        <label for="name">Name:</label>
        <input type="text" id="name" name="name">

        <br><br>

        <label for="country">Country:</label>
        <select id="country" name="country">
            <option>India</option>
            <option>USA</option>
            <option>Canada</option>
            <option>Australia</option>
        </select>

        <br><br>

        <button type="submit">Submit</button>
    </form>

</body>
</html>
````

5: Write short HTML code to show your favorite fruits in a bulleted list and your top 3 programming languages in a numbered list.

````python
%%html
<!DOCTYPE html>
<html>
<head>
    <title>Lists Example</title>
    <style>
        body {
            font-family: Arial, sans-serif;
            margin: 20px;
            background-color: #f0f0f0; /* Light grey background */
        }
        ul {
            list-style-type: square;
            color: green;
        }
        ol {
            list-style-type: upper-roman;
            color: blue;
        }
        h2 {
            color: #333;
        }
        /* Hover effect for list items */
        li {
            padding: 5px;
            margin-bottom: 3px;
            transition: background-color 0.3s ease, color 0.3s ease;
        }
        li:hover {
            background-color: #e0e0e0; /* Slightly darker grey on hover */
            color: #d35400; /* Orange text on hover */
            cursor: pointer; /* Change cursor to pointer on hover */
        }
    </style>
</head>
<body>

    <h2>My Favorite Fruits</h2>
    <ul>
        <li>Apple</li>
        <li>Mango</li>
        <li>Orange</li>
    </ul>

    <h2>Top 3 Programming Languages</h2>
    <ol>
        <li>Python</li>
        <li>Java</li>
        <li>JavaScript</li>
    </ol>

</body>
</html>
````

 7: Here's a form snippet. Can you spot the error and rewrite the correct version?

### Given Code

```html
<form>
    <input name="email">
    <select>
        <option>India</option>
    </select>
    <button>Send</buton>
</form>
```


````python
%%html
<!DOCTYPE html>
<html>
<head>
    <title>Correct Form</title>
</head>
<body>

    <form>
        <input type="email" name="email" placeholder="Enter your email">

        <select name="country">
            <option>India</option>
            <option>USA</option>
            <option>Canada</option>
        </select>

        <br><br>
        <label for="comments">Comments:</label><br>
        <textarea id="comments" name="comments" rows="4" cols="50"></textarea>
        <br><br>

        <button type="submit">Send</button>
    </form>

</body>
</html>
````
